### üìò B- 1: Implement GMF (one component of NCF)

#### Generalized Matrix Factorization (GMF)

**Role**: GMF extends the traditional matrix factorization approach by introducing a flexible way to learn user-item interactions through embeddings. It retains the simplicity of matrix factorization but applies a multiplicative interaction between user and item latent vectors.

<img src="images/GMF.png" width="1000px" />

For explicit feedback (e.g., ratings 1-5): we use ``regression loss``, also known as ``mean squared error loss``:

$$
L_r = \sum_{(u,i) \in \mathcal{Y}} (y_{ui} - \hat{y}_{ui})^2 + \lambda \| \Theta \|^2 
$$

````python
from surprise import AlgoBase
from surprise import Prediction

from tensorflow.keras.layers import Embedding, Input, Dense, Flatten, Concatenate, Multiply
from tensorflow.keras.regularizers import l1, l2
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD, Adagrad, Adam, RMSprop
import numpy as np

class GMF(AlgoBase):
    def __init__(self, n_epochs=20, batch_size=32, embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='explicit'):
        AlgoBase.__init__(self)
        self.n_epochs = n_epochs
        self.batch_size = batch_size
        self.u_embedding_size = embedding_size
        self.i_embedding_size = embedding_size
        self.learning_rate = learning_rate
        self.optimizer = optimizer
        self.feedback_type = feedback_type

    def fit(self, trainset):
            AlgoBase.fit(self, trainset)

            # Create the user and item input layers
            user_input = Input(shape=(1,))
            item_input = Input(shape=(1,))

            # Create the user and item embedding layers
            user_embedding = Embedding(input_dim=trainset.n_users, output_dim=self.u_embedding_size, name='user_embedding',
                                    embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(user_input)
            item_embedding = Embedding(input_dim=trainset.n_items, output_dim=self.i_embedding_size, name='item_embedding',
                                    embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(item_input)

            # Flatten the embedding layers
            user_embedding = Flatten()(user_embedding)
            item_embedding = Flatten()(item_embedding)

            # Element-wise product of user and item embeddings
            product_embeddings = Multiply()([user_embedding, item_embedding])

            # Final prediction layer
            output = Dense(1)(product_embeddings)

            # Create the model
            model = Model(inputs=[user_input, item_input], outputs=output)

            # Compile the model with loss function and optimizer
            loss = 'mean_squared_error'
            if self.optimizer.lower() == "adagrad":
                model.compile(optimizer=Adagrad(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "rmsprop":
                model.compile(optimizer=RMSprop(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "adam":
                model.compile(optimizer=Adam(learning_rate=self.learning_rate), loss=loss)
            else:
                model.compile(optimizer=SGD(learning_rate=self.learning_rate), loss=loss)

            # Train the model
            print('Training GMF...')

            # Access user, item, and rating information from trainset
            user_ids, item_ids, ratings = [], [], []

            for u, i, r in trainset.all_ratings():
                user_ids.append(u)
                item_ids.append(i)
                ratings.append(r)

            # Convert user and item IDs to numpy arrays
            user_ids = np.array(user_ids)
            item_ids = np.array(item_ids)
            ratings = np.array(ratings, dtype=np.float32)

            model.fit([user_ids, item_ids], ratings, epochs=self.n_epochs, batch_size=self.batch_size)
            self.model = model

    def test(self, testset, verbose=False):
        """Test the algorithm on given testset, i.e. estimate all the ratings
        in the given testset.

        Args:
            testset: A test set, as returned by a :ref:`cross-validation
                itertor<use_cross_validation_iterators>` or by the
                :meth:`build_testset() <surprise.Trainset.build_testset>`
                method.
            verbose(bool): Whether to print details for each predictions.
                Default is False.

        Returns:
            A list of :class:`Prediction\
            <surprise.prediction_algorithms.predictions.Prediction>` objects
            that contains all the estimated ratings.
        """

        known_user_ids = []
        known_item_ids = []
        known_ratings = []

        unknown_user_ids = []
        unknown_item_ids = []
        unknown_ratings = []
        unknown_ests = []

        for (u, i, r) in testset:
            try:
                iuid = self.trainset.to_inner_uid(u)
                iiid = self.trainset.to_inner_iid(i)

                known_user_ids.append(iuid)
                known_item_ids.append(iiid)
                known_ratings.append(r)

            except ValueError:
                unknown_user_ids.append("UKN__" + str(u))
                unknown_item_ids.append("UKN__" + str(i))
                unknown_ratings.append(r)
                unknown_ests.append(self.default_prediction())

        # Predict the ratings for all user-item pairs in the testset
        predicted_ratings = self.model.predict([np.array(known_user_ids), np.array(known_item_ids)])

        # Create Prediction objects for each user-item pair
        known_predictions = [
            Prediction(uid=known_user_ids[i], iid=known_item_ids[i], r_ui=known_ratings[i], est=predicted_ratings[i][0], details={})
            for i in range(len(predicted_ratings))
        ]

        unknown_predictions = [
            Prediction(uid=unknown_user_ids[i], iid=unknown_item_ids[i], r_ui=unknown_ratings[i], est=unknown_ests[i], details={})
            for i in range(len(unknown_ratings))
        ]

        predictions = known_predictions + unknown_predictions

        return predictions
````

````python
from surprise import accuracy
import timeit

start = timeit.default_timer()

hyperparameters = {
    'n_epochs':10,
    'embedding_size':10,
    'learning_rate': 0.01,
    'feedback_type': 'explicit'
}

# use the GMF algorithm.
algo_GMF = GMF(**hyperparameters)

# Train the algorithm on the trainset, and predict ratings for the testset
algo_GMF.fit(train_set)
predictions = algo_GMF.test(test_set)

stop = timeit.default_timer()
print('Training Time: ', stop - start)

# Compute accuracy metrics
accuracy.rmse(predictions, verbose=True)
accuracy.mse(predictions, verbose=True)
accuracy.mae(predictions, verbose=True)
````

````python
# Specify a list to store the result
def store_result(cur_result_list, i_method, best_parameter_dict, i_RMSE, i_MSE, i_MAE):
    new_result={
        'method': i_method,
        'best_parameter': best_parameter_dict,
        'RMSE': i_RMSE, 
        'MSE': i_MSE, 
        'MAE': i_MAE, 
    }
    cur_result_list.append(new_result)
    
    return cur_result_list

result_list = []

result_list = store_result(result_list, 'GMF', hyperparameters, accuracy.rmse(predictions, verbose=True), accuracy.mse(predictions, verbose=True), accuracy.mae(predictions, verbose=True))
````

### üìò B- 2: Implement MLP (the other component of NCF)

#### Multi-Layer Perceptron (MLP)

**Role**: MLP is used to learn nonlinear interactions between users and items. It is more expressive than GMF, allowing the model to capture intricate and complex relationships.

<img src="images/MLP.png" width="1000px" />

For explicit feedback (e.g., ratings 1-5): we use ``regression loss``, also known as ``mean squared error loss``:

$$
L_r = \sum_{(u,i) \in \mathcal{Y}} (y_{ui} - \hat{y}_{ui})^2 + \lambda \| \Theta \|^2 
$$

````python
class MLP(AlgoBase):
    """
    'MLP_layers': 
        the array size, i.e., 2, denotes the number of layers; 
        each value denotes the number of neurons at the corresponding layer (i.e., 20 neurons at the 1st layer and 10 neurons at the 2nd layer).
    """
    def __init__(self, MLP_layers=[20,10], n_epochs=20, batch_size=32, embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='explicit'):
        AlgoBase.__init__(self)
        self.n_epochs = n_epochs
        self.batch_size = batch_size
        self.u_embedding_size = embedding_size
        self.i_embedding_size = embedding_size
        self.learning_rate = learning_rate
        self.optimizer = optimizer
        self.MLP_layers = MLP_layers
        self.feedback_type = feedback_type

    def fit(self, trainset):
            AlgoBase.fit(self, trainset)

            num_layer = len(self.MLP_layers)

            # Create the user and item input layers
            user_input = Input(shape=(1,))
            item_input = Input(shape=(1,))

            # Create the user and item embedding layers
            user_embedding = Embedding(input_dim=trainset.n_users, output_dim=self.u_embedding_size, name='user_embedding',
                                    embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(user_input)
            item_embedding = Embedding(input_dim=trainset.n_items, output_dim=self.i_embedding_size, name='item_embedding',
                                    embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(item_input)

            # Flatten the embedding layers
            user_embedding = Flatten()(user_embedding)
            item_embedding = Flatten()(item_embedding)

            # Concatenate the user and item embeddings
            merged_embeddings = Concatenate()([user_embedding, item_embedding])

            # MLP layers
            for idx in range(num_layer):
                layer = Dense(self.MLP_layers[idx], kernel_regularizer = l2(0.0), activation='relu', name='layer%d' %idx)
                merged_embeddings = layer(merged_embeddings)

            # Final prediction layer
            output = Dense(1)(merged_embeddings)

            # Create the model
            model = Model(inputs=[user_input, item_input], outputs=output)

            # Compile the model with optimizer
            loss = 'mean_squared_error'
            if self.optimizer.lower() == "adagrad":
                model.compile(optimizer=Adagrad(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "rmsprop":
                model.compile(optimizer=RMSprop(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "adam":
                model.compile(optimizer=Adam(learning_rate=self.learning_rate), loss=loss)
            else:
                model.compile(optimizer=SGD(learning_rate=self.learning_rate), loss=loss)

            # Train the model
            print('Training MLP...')

            # Access user, item, and rating information from trainset
            user_ids, item_ids, ratings = [], [], []

            for u, i, r in trainset.all_ratings():
                user_ids.append(u)
                item_ids.append(i)
                ratings.append(r)

            # Convert user and item IDs to numpy arrays
            user_ids = np.array(user_ids)
            item_ids = np.array(item_ids)
            ratings = np.array(ratings, dtype=np.float32)

            model.fit([user_ids, item_ids], ratings, epochs=self.n_epochs, batch_size=self.batch_size)
            self.model = model

    def test(self, testset, verbose=False):
        """Test the algorithm on given testset, i.e. estimate all the ratings
        in the given testset.

        Args:
            testset: A test set, as returned by a :ref:`cross-validation
                itertor<use_cross_validation_iterators>` or by the
                :meth:`build_testset() <surprise.Trainset.build_testset>`
                method.
            verbose(bool): Whether to print details for each predictions.
                Default is False.

        Returns:
            A list of :class:`Prediction\
            <surprise.prediction_algorithms.predictions.Prediction>` objects
            that contains all the estimated ratings.
        """

        known_user_ids = []
        known_item_ids = []
        known_ratings = []

        unknown_user_ids = []
        unknown_item_ids = []
        unknown_ratings = []
        unknown_ests = []

        for (u, i, r) in testset:
            try:
                iuid = self.trainset.to_inner_uid(u)
                iiid = self.trainset.to_inner_iid(i)

                known_user_ids.append(iuid)
                known_item_ids.append(iiid)
                known_ratings.append(r)

            except ValueError:
                unknown_user_ids.append("UKN__" + str(u))
                unknown_item_ids.append("UKN__" + str(i))
                unknown_ratings.append(r)
                unknown_ests.append(self.default_prediction())

        # Predict the ratings for all user-item pairs in the testset
        predicted_ratings = self.model.predict([np.array(known_user_ids), np.array(known_item_ids)])

        # Create Prediction objects for each user-item pair
        known_predictions = [
            Prediction(uid=known_user_ids[i], iid=known_item_ids[i], r_ui=known_ratings[i], est=predicted_ratings[i][0], details={})
            for i in range(len(predicted_ratings))
        ]

        unknown_predictions = [
            Prediction(uid=unknown_user_ids[i], iid=unknown_item_ids[i], r_ui=unknown_ratings[i], est=unknown_ests[i], details={})
            for i in range(len(unknown_ratings))
        ]

        predictions = known_predictions + unknown_predictions

        return predictions
````

````python
from surprise import accuracy
import timeit

start = timeit.default_timer()

hyperparameters = {
    'n_epochs':20,
    'MLP_layers': [20, 10],
    'embedding_size':10,
    'learning_rate': 0.01,
    'feedback_type': 'explicit'
}

# use the MLP algorithm.
algo_MLP = MLP(**hyperparameters)

# Train the algorithm on the trainset, and predict ratings for the testset
algo_MLP.fit(train_set)
predictions = algo_MLP.test(test_set)

stop = timeit.default_timer()
print('Training Time: ', stop - start)

# Compute accuracy metrics
accuracy.rmse(predictions, verbose=True)
accuracy.mse(predictions, verbose=True)
accuracy.mae(predictions, verbose=True)

result_list = store_result(result_list, 'MLP', hyperparameters, accuracy.rmse(predictions, verbose=True), accuracy.mse(predictions, verbose=True), accuracy.mae(predictions, verbose=True))
````

### üìò B- 3: Fusion of GMF and MLP

#### Neural Matrix Factorization (NeuMF) Layer

**Role**: This layer combines the outputs from GMF and MLP to form the final recommendation prediction. 

<img src="images/NeuMF.png" width="1000px" />


For explicit feedback (e.g., ratings 1-5): we use ``regression loss``, also known as ``mean squared error loss``:

$$
L_r = \sum_{(u,i) \in \mathcal{Y}} (y_{ui} - \hat{y}_{ui})^2 + \lambda \| \Theta \|^2 
$$

````python
class NCF(AlgoBase):
    def __init__(self, MLP_layers=[20,10], n_epochs=20, batch_size=32, embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='explicit'):
        AlgoBase.__init__(self)
        self.n_epochs = n_epochs
        self.batch_size = batch_size
        self.u_embedding_size = embedding_size
        self.i_embedding_size = embedding_size
        self.learning_rate = learning_rate
        self.optimizer = optimizer
        self.MLP_layers = MLP_layers
        self.feedback_type = feedback_type

    def fit(self, trainset):
            AlgoBase.fit(self, trainset)

            # Create the user and item input layers
            user_input = Input(shape=(1,))
            item_input = Input(shape=(1,))

            MF_Embedding_User = Embedding(input_dim = trainset.n_users, output_dim = self.u_embedding_size, name = 'mf_embedding_user',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))
            MF_Embedding_Item = Embedding(input_dim = trainset.n_items, output_dim = self.i_embedding_size, name = 'mf_embedding_item',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))

            MLP_Embedding_User = Embedding(input_dim = trainset.n_users, output_dim = self.u_embedding_size, name = "mlp_embedding_user",
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))
            MLP_Embedding_Item = Embedding(input_dim = trainset.n_items, output_dim = self.i_embedding_size, name = 'mlp_embedding_item',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))

            # MF part
            mf_user_latent = Flatten()(MF_Embedding_User(user_input))
            mf_item_latent = Flatten()(MF_Embedding_Item(item_input))
            mf_vector = Multiply()([mf_user_latent, mf_item_latent]) # element-wise multiply

            # MLP part
            mlp_user_latent = Flatten()(MLP_Embedding_User(user_input))
            mlp_item_latent = Flatten()(MLP_Embedding_Item(item_input))
            mlp_vector = Concatenate()([mlp_user_latent, mlp_item_latent])
            for idx in range(len(self.MLP_layers)):
                layer = Dense(self.MLP_layers[idx], kernel_regularizer= l2(0.0), activation='relu', name="layer%d" %idx)
                mlp_vector = layer(mlp_vector)
            
            # Concatenate MF and MLP parts
            predict_vector = Concatenate()([mf_vector, mlp_vector])

            prediction = Dense(1)(predict_vector)

            model = Model(inputs=[user_input, item_input], outputs=prediction)

            # Compile the model with optimizer
            loss = 'mean_squared_error'
            if self.optimizer.lower() == "adagrad":
                model.compile(optimizer=Adagrad(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "rmsprop":
                model.compile(optimizer=RMSprop(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "adam":
                model.compile(optimizer=Adam(learning_rate=self.learning_rate), loss=loss)
            else:
                model.compile(optimizer=SGD(learning_rate=self.learning_rate), loss=loss)

            # Train the model
            print('Training NCF...')

            # Access user, item, and rating information from trainset
            user_ids, item_ids, ratings = [], [], []

            for u, i, r in trainset.all_ratings():
                user_ids.append(u)
                item_ids.append(i)
                ratings.append(r)

            # Convert user and item IDs to numpy arrays
            user_ids = np.array(user_ids)
            item_ids = np.array(item_ids)
            ratings = np.array(ratings, dtype=np.float32)

            model.fit([user_ids, item_ids], ratings, epochs=self.n_epochs, batch_size=self.batch_size)
            self.model = model
    
    def test(self, testset, verbose=False):
        """Test the algorithm on given testset, i.e. estimate all the ratings
        in the given testset.

        Args:
            testset: A test set, as returned by a :ref:`cross-validation
                itertor<use_cross_validation_iterators>` or by the
                :meth:`build_testset() <surprise.Trainset.build_testset>`
                method.
            verbose(bool): Whether to print details for each predictions.
                Default is False.

        Returns:
            A list of :class:`Prediction\
            <surprise.prediction_algorithms.predictions.Prediction>` objects
            that contains all the estimated ratings.
        """

        known_user_ids = []
        known_item_ids = []
        known_ratings = []

        unknown_user_ids = []
        unknown_item_ids = []
        unknown_ratings = []
        unknown_ests = []

        for (u, i, r) in testset:
            try:
                iuid = self.trainset.to_inner_uid(u)
                iiid = self.trainset.to_inner_iid(i)

                known_user_ids.append(iuid)
                known_item_ids.append(iiid)
                known_ratings.append(r)

            except ValueError:
                unknown_user_ids.append("UKN__" + str(u))
                unknown_item_ids.append("UKN__" + str(i))
                unknown_ratings.append(r)
                unknown_ests.append(self.default_prediction())

        # Predict the ratings for all user-item pairs in the testset
        predicted_ratings = self.model.predict([np.array(known_user_ids), np.array(known_item_ids)])

        # Create Prediction objects for each user-item pair
        known_predictions = [
            Prediction(uid=known_user_ids[i], iid=known_item_ids[i], r_ui=known_ratings[i], est=predicted_ratings[i][0], details={})
            for i in range(len(predicted_ratings))
        ]

        unknown_predictions = [
            Prediction(uid=unknown_user_ids[i], iid=unknown_item_ids[i], r_ui=unknown_ratings[i], est=unknown_ests[i], details={})
            for i in range(len(unknown_ratings))
        ]

        predictions = known_predictions + unknown_predictions

        return predictions
````

### üìò B- 4: Pre-training

<img src="images/PreTrain.png" width="1200px" />

In machine learning, pre-training is to initialize models with parameters learned from a related but separate task, which can help improve the final model's performance. This technique may assist the further fine-tuning on the primary task, enhancing the model's ability to capture complex patterns. 

In the context of Neural Collaborative Filtering (NCF), the pre-training process involves independently training the Generalized Matrix Factorization (GMF) and Multi-Layer Perceptron (MLP) components before integrating them into the complete NCF architecture. By allowing each component to learn its own specialized features independently during the pre-training phase, the method ensures that when integrated, the components are already well-aligned with their respective tasks. This leads to a more cohesive and effective overall model when the components are finally combined and fine-tuned together.

````python
class NCF(AlgoBase):
    def __init__(self, MLP_layers=[20,10], n_epochs=20, n_epochs_GMF=5, n_epochs_MLP=20, batch_size=32, embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='explicit', pretrain=False):
        AlgoBase.__init__(self)
        self.n_epochs = n_epochs
        self.n_epochs_GMF = n_epochs_GMF
        self.n_epochs_MLP = n_epochs_MLP
        self.batch_size = batch_size
        self.u_embedding_size = embedding_size
        self.i_embedding_size = embedding_size
        self.learning_rate = learning_rate
        self.optimizer = optimizer
        self.MLP_layers = MLP_layers
        self.feedback_type = feedback_type
        self.pretrain = pretrain

    def GMF_pretrain(self, trainset, n_epochs=5, batch_size=32, u_embedding_size=10, i_embedding_size=10, learning_rate=0.01, optimizer='sgd'):
        # Access user, item, and rating information from trainset
        user_ids, item_ids, ratings = [], [], []

        for u, i, r in trainset.all_ratings():
            user_ids.append(u)
            item_ids.append(i)
            ratings.append(r)

        # Convert user and item IDs to numpy arrays
        user_ids = np.array(user_ids)
        item_ids = np.array(item_ids)
        ratings = np.array(ratings, dtype=np.float32)

        # Create the user and item input layers
        user_input = Input(shape=(1,))
        item_input = Input(shape=(1,))

        # Create the user and item embedding layers
        user_embedding = Embedding(input_dim=trainset.n_users, output_dim=u_embedding_size, name='user_embedding',
                                  embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(user_input)
        item_embedding = Embedding(input_dim=trainset.n_items, output_dim=i_embedding_size, name='item_embedding',
                                  embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(item_input)

        # Flatten the embedding layers
        user_embedding = Flatten()(user_embedding)
        item_embedding = Flatten()(item_embedding)

        # Element-wise product of user and item embeddings
        product_embeddings = Multiply()([user_embedding, item_embedding])

        # Final prediction layer
        output = Dense(1)(product_embeddings)

        # Create the model
        model = Model(inputs=[user_input, item_input], outputs=output)

        # Compile the model with optimizer
        loss = 'mean_squared_error'
        if optimizer.lower() == "adagrad":
            model.compile(optimizer=Adagrad(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "rmsprop":
            model.compile(optimizer=RMSprop(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "adam":
            model.compile(optimizer=Adam(learning_rate=learning_rate), loss=loss)
        else:
            model.compile(optimizer=SGD(learning_rate=learning_rate), loss=loss)

        # Train the model
        model.fit([user_ids, item_ids], ratings, epochs=n_epochs, batch_size=batch_size)
        return model
        

    def MLP_pretrain(self, trainset, n_epochs=20, batch_size=32, u_embedding_size=10, i_embedding_size=10, learning_rate=0.01, optimizer='sgd', MLP_layers=[20,10]):
        num_layer = len(MLP_layers)

        # Access user, item, and rating information from trainset
        user_ids, item_ids, ratings = [], [], []

        for u, i, r in trainset.all_ratings():
            user_ids.append(u)
            item_ids.append(i)
            ratings.append(r)

        # Convert user and item IDs to numpy arrays
        user_ids = np.array(user_ids)
        item_ids = np.array(item_ids)
        ratings = np.array(ratings, dtype=np.float32)

        # Create the user and item input layers
        user_input = Input(shape=(1,))
        item_input = Input(shape=(1,))

        # Create the user and item embedding layers
        user_embedding = Embedding(input_dim=trainset.n_users, output_dim=u_embedding_size, name='user_embedding',
                                   embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(user_input)
        item_embedding = Embedding(input_dim=trainset.n_items, output_dim=i_embedding_size, name='item_embedding',
                                   embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(item_input)

        # Flatten the embedding layers
        user_embedding = Flatten()(user_embedding)
        item_embedding = Flatten()(item_embedding)

        # Concatenate the user and item embeddings
        merged_embeddings = Concatenate()([user_embedding, item_embedding])

        # MLP layers
        for idx in range(num_layer):
            layer = Dense(MLP_layers[idx], kernel_regularizer = l2(0.0), activation='relu', name='layer%d' %idx)
            merged_embeddings = layer(merged_embeddings)

        # Final prediction layer
        output = Dense(1)(merged_embeddings)


        # Create the model
        model = Model(inputs=[user_input, item_input], outputs=output)

        # Compile the model with optimizer
        loss = 'mean_squared_error'
        if optimizer.lower() == "adagrad":
            model.compile(optimizer=Adagrad(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "rmsprop":
            model.compile(optimizer=RMSprop(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "adam":
            model.compile(optimizer=Adam(learning_rate=learning_rate), loss=loss)
        else:
            model.compile(optimizer=SGD(learning_rate=learning_rate), loss=loss)

        # Train the model
        model.fit([user_ids, item_ids], ratings, epochs=n_epochs, batch_size=batch_size)
        return model

    def fit(self, trainset):
            AlgoBase.fit(self, trainset)

            # Create the user and item input layers
            user_input = Input(shape=(1,))
            item_input = Input(shape=(1,))

            MF_Embedding_User = Embedding(input_dim = trainset.n_users, output_dim = self.u_embedding_size, name = 'mf_embedding_user',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))
            MF_Embedding_Item = Embedding(input_dim = trainset.n_items, output_dim = self.i_embedding_size, name = 'mf_embedding_item',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))

            MLP_Embedding_User = Embedding(input_dim = trainset.n_users, output_dim = self.u_embedding_size, name = "mlp_embedding_user",
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))
            MLP_Embedding_Item = Embedding(input_dim = trainset.n_items, output_dim = self.i_embedding_size, name = 'mlp_embedding_item',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))

            # MF part
            mf_user_latent = Flatten()(MF_Embedding_User(user_input))
            mf_item_latent = Flatten()(MF_Embedding_Item(item_input))
            mf_vector = Multiply()([mf_user_latent, mf_item_latent]) # element-wise multiply

            # MLP part
            mlp_user_latent = Flatten()(MLP_Embedding_User(user_input))
            mlp_item_latent = Flatten()(MLP_Embedding_Item(item_input))
            mlp_vector = Concatenate()([mlp_user_latent, mlp_item_latent])
            for idx in range(len(self.MLP_layers)):
                layer = Dense(self.MLP_layers[idx], kernel_regularizer= l2(0.0), activation='relu', name="layer%d" %idx)
                mlp_vector = layer(mlp_vector)
            
            # Concatenate MF and MLP parts
            predict_vector = Concatenate()([mf_vector, mlp_vector])


            prediction = Dense(1)(predict_vector)

            model = Model(inputs=[user_input, item_input], outputs=prediction)

            if self.pretrain:
                print('Pretraining GMF...')
                gmf_model = self.GMF_pretrain(trainset, self.n_epochs_GMF, self.batch_size, self.u_embedding_size, self.i_embedding_size, self.learning_rate, self.optimizer)
                print('Pretraining MLP...')
                mlp_model = self.MLP_pretrain(trainset, self.n_epochs_MLP, self.batch_size, self.u_embedding_size, self.i_embedding_size, self.learning_rate, self.optimizer, self.MLP_layers)

                # MF embeddings
                gmf_user_embeddings = gmf_model.get_layer('user_embedding').get_weights()
                gmf_item_embeddings = gmf_model.get_layer('item_embedding').get_weights()
                model.get_layer('mf_embedding_user').set_weights(gmf_user_embeddings)
                model.get_layer('mf_embedding_item').set_weights(gmf_item_embeddings)

                # MLP embeddings
                mlp_user_embeddings = mlp_model.get_layer('user_embedding').get_weights()
                mlp_item_embeddings = mlp_model.get_layer('item_embedding').get_weights()
                model.get_layer('mlp_embedding_user').set_weights(mlp_user_embeddings)
                model.get_layer('mlp_embedding_item').set_weights(mlp_item_embeddings)

            # Compile the model with optimizer
            loss = 'mean_squared_error'
            if self.optimizer.lower() == "adagrad":
                model.compile(optimizer=Adagrad(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "rmsprop":
                model.compile(optimizer=RMSprop(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "adam":
                model.compile(optimizer=Adam(learning_rate=self.learning_rate), loss=loss)
            else:
                model.compile(optimizer=SGD(learning_rate=self.learning_rate), loss=loss)

            # Train the model
            print('Training NCF...')

            # Access user, item, and rating information from trainset
            user_ids, item_ids, ratings = [], [], []

            for u, i, r in trainset.all_ratings():
                user_ids.append(u)
                item_ids.append(i)
                ratings.append(r)

            # Convert user and item IDs to numpy arrays
            user_ids = np.array(user_ids)
            item_ids = np.array(item_ids)
            ratings = np.array(ratings, dtype=np.float32)
                        
            model.fit([user_ids, item_ids], ratings, epochs=self.n_epochs, batch_size=self.batch_size)
            self.model = model
    
    def test(self, testset, verbose=False):
        """Test the algorithm on given testset, i.e. estimate all the ratings
        in the given testset.

        Args:
            testset: A test set, as returned by a :ref:`cross-validation
                itertor<use_cross_validation_iterators>` or by the
                :meth:`build_testset() <surprise.Trainset.build_testset>`
                method.
            verbose(bool): Whether to print details for each predictions.
                Default is False.

        Returns:
            A list of :class:`Prediction\
            <surprise.prediction_algorithms.predictions.Prediction>` objects
            that contains all the estimated ratings.
        """

        known_user_ids = []
        known_item_ids = []
        known_ratings = []

        unknown_user_ids = []
        unknown_item_ids = []
        unknown_ratings = []
        unknown_ests = []

        for (u, i, r) in testset:
            try:
                iuid = self.trainset.to_inner_uid(u)
                iiid = self.trainset.to_inner_iid(i)

                known_user_ids.append(iuid)
                known_item_ids.append(iiid)
                known_ratings.append(r)

            except ValueError:
                unknown_user_ids.append("UKN__" + str(u))
                unknown_item_ids.append("UKN__" + str(i))
                unknown_ratings.append(r)
                unknown_ests.append(self.default_prediction())

        # Predict the ratings for all user-item pairs in the testset
        predicted_ratings = self.model.predict([np.array(known_user_ids), np.array(known_item_ids)])

        # Create Prediction objects for each user-item pair
        known_predictions = [
            Prediction(uid=known_user_ids[i], iid=known_item_ids[i], r_ui=known_ratings[i], est=predicted_ratings[i][0], details={})
            for i in range(len(predicted_ratings))
        ]

        unknown_predictions = [
            Prediction(uid=unknown_user_ids[i], iid=unknown_item_ids[i], r_ui=unknown_ratings[i], est=unknown_ests[i], details={})
            for i in range(len(unknown_ratings))
        ]

        predictions = known_predictions + unknown_predictions

        return predictions
````

**Thinking**: Can we train a recommendation model using only data like this?

Implicit feedback only reveals what a user has interacted with, not what they have disliked. lt lacks information about items that a user might not be interested in.

Negative sampling introduces examples of items the user hasn't interacted with. This can help the model learn to differentiate between preferred items and items the user is less likely to engage with. lt essentially teaches the model what not to recommend.

<img src="images/NS.png" width="1000px" />

In implicit feedback, each observed interaction normally indicates a user‚Äôs interest in an item and leads to a positive training sample. As for negative training samples, a widely adopted approach is to randomly select some uninteracted items for a user. 

````python
import pandas as pd
import numpy as np


def negative_sampling(df, num_neg_samples=1):
    """
    Generates negative samples (items not interacted with by the user)

    Args:
        df: The original DataFrame.
        num_neg_samples: Number of negative samples to generate per user-item interaction.

    Returns:
        DataFrame with negative samples added.
    """
    all_items = set(df['item'].unique())
    df_neg = []

    for user in df['user'].unique():
        interacted_items = set(df[df['user'] == user]['item'])
        non_interacted_items = list(all_items - interacted_items)

        for item in df[df['user'] == user]['item']: # Loop through actual rated items
            negative_samples = np.random.choice(non_interacted_items, size=num_neg_samples, replace=False) # replace=False will avoid generating the same sample multiple times in one go.
            for neg_item in negative_samples:
                df_neg.append([user, neg_item, 0, 0])  # Rating and timestamp set to 0

    df_neg = pd.DataFrame(df_neg, columns=["user", "item", "rating", "timestamp"])
    df_final = pd.concat([df, df_neg], ignore_index=True)
    return df_final

# Example usage, generating 1 negative sample per positive interaction:
df_with_negative_samples = negative_sampling(df)
df_with_negative_samples
````

Train NCF model using the implicit feedback

For implicit feedback (e.g., watches 0/1): we use ``classification loss``, also known as ``binary cross entropy loss``:

$$
L_c = -\left( \sum_{(u,i) \in \mathcal{Y} \cup \mathcal{Y}^-} y_{ui} \log \hat{y}_{ui} + (1 - y_{ui}) \log (1 - \hat{y}_{ui}) \right) + \lambda \| \Theta \|^2 
$$

````python
class NCF(AlgoBase):
    def __init__(self, MLP_layers=[20,10], n_epochs=20, n_epochs_GMF=5, n_epochs_MLP=20, batch_size=32, embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='implicit', pretrain=False):
        AlgoBase.__init__(self)
        self.n_epochs = n_epochs
        self.n_epochs_GMF = n_epochs_GMF
        self.n_epochs_MLP = n_epochs_MLP
        self.batch_size = batch_size
        self.u_embedding_size = embedding_size
        self.i_embedding_size = embedding_size
        self.learning_rate = learning_rate
        self.optimizer = optimizer
        self.feedback_type = feedback_type
        self.MLP_layers = MLP_layers
        self.pretrain = pretrain

    def GMF_pretrain(self, trainset, n_epochs=5, batch_size=32, u_embedding_size=10, i_embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='implicit'):
        # Access user, item, and rating information from trainset
        user_ids, item_ids, ratings = [], [], []

        for u, i, r in trainset.all_ratings():
            user_ids.append(u)
            item_ids.append(i)
            ratings.append(r)

        # Convert user and item IDs to numpy arrays
        user_ids = np.array(user_ids)
        item_ids = np.array(item_ids)
        ratings = np.array(ratings, dtype=np.float32)

        # Create the user and item input layers
        user_input = Input(shape=(1,))
        item_input = Input(shape=(1,))

        # Create the user and item embedding layers
        user_embedding = Embedding(input_dim=trainset.n_users, output_dim=u_embedding_size, name='user_embedding',
                                  embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(user_input)
        item_embedding = Embedding(input_dim=trainset.n_items, output_dim=i_embedding_size, name='item_embedding',
                                  embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(item_input)

        # Flatten the embedding layers
        user_embedding = Flatten()(user_embedding)
        item_embedding = Flatten()(item_embedding)

        # Element-wise product of user and item embeddings
        product_embeddings = Multiply()([user_embedding, item_embedding])

        # Final prediction layer
        if feedback_type == 'explicit':
            output = Dense(1)(product_embeddings)
        elif feedback_type == 'implicit':
            # Add an activation function
            output = Dense(1, activation='sigmoid')(product_embeddings)

        # Create the model
        model = Model(inputs=[user_input, item_input], outputs=output)

        # Compile the model with optimizer
        loss = 'mean_squared_error' if feedback_type == 'explicit' else 'binary_crossentropy'
        if optimizer.lower() == "adagrad":
            model.compile(optimizer=Adagrad(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "rmsprop":
            model.compile(optimizer=RMSprop(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "adam":
            model.compile(optimizer=Adam(learning_rate=learning_rate), loss=loss)
        else:
            model.compile(optimizer=SGD(learning_rate=learning_rate), loss=loss)

        # Train the model
        model.fit([user_ids, item_ids], ratings, epochs=n_epochs, batch_size=batch_size)
        return model
        

    def MLP_pretrain(self, trainset, n_epochs=20, batch_size=32, u_embedding_size=10, i_embedding_size=10, learning_rate=0.01, optimizer='sgd', feedback_type='implicit', MLP_layers=[20,10]):
        num_layer = len(MLP_layers)

        # Access user, item, and rating information from trainset
        user_ids, item_ids, ratings = [], [], []

        for u, i, r in trainset.all_ratings():
            user_ids.append(u)
            item_ids.append(i)
            ratings.append(r)

        # Convert user and item IDs to numpy arrays
        user_ids = np.array(user_ids)
        item_ids = np.array(item_ids)
        ratings = np.array(ratings, dtype=np.float32)

        # Create the user and item input layers
        user_input = Input(shape=(1,))
        item_input = Input(shape=(1,))

        # Create the user and item embedding layers
        user_embedding = Embedding(input_dim=trainset.n_users, output_dim=u_embedding_size, name='user_embedding',
                                   embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(user_input)
        item_embedding = Embedding(input_dim=trainset.n_items, output_dim=i_embedding_size, name='item_embedding',
                                   embeddings_initializer='random_normal', embeddings_regularizer=l2(0.0))(item_input)

        # Flatten the embedding layers
        user_embedding = Flatten()(user_embedding)
        item_embedding = Flatten()(item_embedding)

        # Concatenate the user and item embeddings
        merged_embeddings = Concatenate()([user_embedding, item_embedding])

        # MLP layers
        for idx in range(num_layer):
            layer = Dense(MLP_layers[idx], kernel_regularizer = l2(0.0), activation='relu', name='layer%d' %idx)
            merged_embeddings = layer(merged_embeddings)

        # Final prediction layer
        if feedback_type == 'explicit':
            output = Dense(1)(merged_embeddings)
        elif feedback_type == 'implicit':
            # Add activation function
            output = Dense(1, activation='sigmoid')(merged_embeddings)

        # Create the model
        model = Model(inputs=[user_input, item_input], outputs=output)

        # Compile the model with optimizer
        loss = 'mean_squared_error' if feedback_type == 'explicit' else 'binary_crossentropy'
        if optimizer.lower() == "adagrad":
            model.compile(optimizer=Adagrad(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "rmsprop":
            model.compile(optimizer=RMSprop(learning_rate=learning_rate), loss=loss)
        elif optimizer.lower() == "adam":
            model.compile(optimizer=Adam(learning_rate=learning_rate), loss=loss)
        else:
            model.compile(optimizer=SGD(learning_rate=learning_rate), loss=loss)

        # Train the model
        model.fit([user_ids, item_ids], ratings, epochs=n_epochs, batch_size=batch_size)
        return model

    def fit(self, trainset):
            AlgoBase.fit(self, trainset)

            # Create the user and item input layers
            user_input = Input(shape=(1,))
            item_input = Input(shape=(1,))

            MF_Embedding_User = Embedding(input_dim = trainset.n_users, output_dim = self.u_embedding_size, name = 'mf_embedding_user',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))
            MF_Embedding_Item = Embedding(input_dim = trainset.n_items, output_dim = self.i_embedding_size, name = 'mf_embedding_item',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))

            MLP_Embedding_User = Embedding(input_dim = trainset.n_users, output_dim = self.u_embedding_size, name = "mlp_embedding_user",
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))
            MLP_Embedding_Item = Embedding(input_dim = trainset.n_items, output_dim = self.i_embedding_size, name = 'mlp_embedding_item',
                                        embeddings_initializer = 'random_normal', embeddings_regularizer = l2(0.0))

            # MF part
            mf_user_latent = Flatten()(MF_Embedding_User(user_input))
            mf_item_latent = Flatten()(MF_Embedding_Item(item_input))
            mf_vector = Multiply()([mf_user_latent, mf_item_latent]) # element-wise multiply

            # MLP part
            mlp_user_latent = Flatten()(MLP_Embedding_User(user_input))
            mlp_item_latent = Flatten()(MLP_Embedding_Item(item_input))
            mlp_vector = Concatenate()([mlp_user_latent, mlp_item_latent])
            for idx in range(len(self.MLP_layers)):
                layer = Dense(self.MLP_layers[idx], kernel_regularizer= l2(0.0), activation='relu', name="layer%d" %idx)
                mlp_vector = layer(mlp_vector)
            
            # Concatenate MF and MLP parts
            predict_vector = Concatenate()([mf_vector, mlp_vector])

            if self.feedback_type == 'explicit':
                prediction = Dense(1)(predict_vector)
            elif self.feedback_type == 'implicit':
                # Add activation function
                prediction = Dense(1, activation='sigmoid')(predict_vector)

            model = Model(inputs=[user_input, item_input], outputs=prediction)

            if self.pretrain:
                print('Pretraining GMF...')
                gmf_model = self.GMF_pretrain(trainset, self.n_epochs_GMF, self.batch_size, self.u_embedding_size, self.i_embedding_size, self.learning_rate, self.optimizer, self.feedback_type)
                print('Pretraining MLP...')
                mlp_model = self.MLP_pretrain(trainset, self.n_epochs_MLP, self.batch_size, self.u_embedding_size, self.i_embedding_size, self.learning_rate, self.optimizer, self.feedback_type, self.MLP_layers)

                # MF embeddings
                gmf_user_embeddings = gmf_model.get_layer('user_embedding').get_weights()
                gmf_item_embeddings = gmf_model.get_layer('item_embedding').get_weights()
                model.get_layer('mf_embedding_user').set_weights(gmf_user_embeddings)
                model.get_layer('mf_embedding_item').set_weights(gmf_item_embeddings)

                # MLP embeddings
                mlp_user_embeddings = mlp_model.get_layer('user_embedding').get_weights()
                mlp_item_embeddings = mlp_model.get_layer('item_embedding').get_weights()
                model.get_layer('mlp_embedding_user').set_weights(mlp_user_embeddings)
                model.get_layer('mlp_embedding_item').set_weights(mlp_item_embeddings)

            # Compile the model with optimizer
            loss = 'mean_squared_error' if self.feedback_type == 'explicit' else 'binary_crossentropy'
            if self.optimizer.lower() == "adagrad":
                model.compile(optimizer=Adagrad(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "rmsprop":
                model.compile(optimizer=RMSprop(learning_rate=self.learning_rate), loss=loss)
            elif self.optimizer.lower() == "adam":
                model.compile(optimizer=Adam(learning_rate=self.learning_rate), loss=loss)
            else:
                model.compile(optimizer=SGD(learning_rate=self.learning_rate), loss=loss)

            # Train the model
            print('Training NCF...')

            # Access user, item, and rating information from trainset
            user_ids, item_ids, ratings = [], [], []

            for u, i, r in trainset.all_ratings():
                user_ids.append(u)
                item_ids.append(i)
                ratings.append(r)

            # Convert user and item IDs to numpy arrays
            user_ids = np.array(user_ids)
            item_ids = np.array(item_ids)
            ratings = np.array(ratings, dtype=np.float32)
                        
            model.fit([user_ids, item_ids], ratings, epochs=self.n_epochs, batch_size=self.batch_size)
            self.model = model
    
    def test(self, testset, verbose=False):
        """Test the algorithm on given testset, i.e. estimate all the ratings
        in the given testset.

        Args:
            testset: A test set, as returned by a :ref:`cross-validation
                itertor<use_cross_validation_iterators>` or by the
                :meth:`build_testset() <surprise.Trainset.build_testset>`
                method.
            verbose(bool): Whether to print details for each predictions.
                Default is False.

        Returns:
            A list of :class:`Prediction\
            <surprise.prediction_algorithms.predictions.Prediction>` objects
            that contains all the estimated ratings.
        """

        known_user_ids = []
        known_item_ids = []
        known_ratings = []

        unknown_user_ids = []
        unknown_item_ids = []
        unknown_ratings = []
        unknown_ests = []

        for (u, i, r) in testset:
            try:
                iuid = self.trainset.to_inner_uid(u)
                iiid = self.trainset.to_inner_iid(i)

                known_user_ids.append(iuid)
                known_item_ids.append(iiid)
                known_ratings.append(r)

            except ValueError:
                unknown_user_ids.append("UKN__" + str(u))
                unknown_item_ids.append("UKN__" + str(i))
                unknown_ratings.append(r)
                unknown_ests.append(self.default_prediction())

        # Predict the ratings for all user-item pairs in the testset
        predicted_ratings = self.model.predict([np.array(known_user_ids), np.array(known_item_ids)])

        # Create Prediction objects for each user-item pair
        known_predictions = [
            Prediction(uid=known_user_ids[i], iid=known_item_ids[i], r_ui=known_ratings[i], est=predicted_ratings[i][0], details={})
            for i in range(len(predicted_ratings))
        ]

        unknown_predictions = [
            Prediction(uid=unknown_user_ids[i], iid=unknown_item_ids[i], r_ui=unknown_ratings[i], est=unknown_ests[i], details={})
            for i in range(len(unknown_ratings))
        ]

        predictions = known_predictions + unknown_predictions

        return predictions
````
